```python
import pandas as pd
```

```python
case_url = 'https://datacatalog.cookcountyil.gov/api/views/cjeq-bs86/rows.csv?accessType=DOWNLOAD'
cremation_url = 'https://datacatalog.cookcountyil.gov/api/views/f9wk-2wb9/rows.csv?accessType=DOWNLOAD'

cases_df = pd.read_csv(case_url)
cremations_df = pd.read_csv(cremation_url)
```

```python
# Standardize the shared case number column
cremations_df.rename(columns={'Case': 'case_number'}, inplace=True)
cases_df.rename(columns={'Case Number': 'case_number'}, inplace=True)

# Strip whitespace and ensure consistent formatting
for df in [cremations_df, cases_df]:
    df['case_number'] = df['case_number'].astype(str).str.strip().str.upper()
```

```python
merged_df = pd.merge(cremations_df, cases_df, on='case_number', how='inner')
```
